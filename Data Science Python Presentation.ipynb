```python
# Importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assigning file name to object and importing excel into pandas dataframe
file = 'WDIW Dataset.xlsx'
df_world = pd.read_excel(file, sheet_name = 'Raw' )

# subsetting assigned region and define country names as index
df_aus = df_world[df_world['Cool Name'] == 'Cliffhanger']
df_aus.index = df_aus['Country Name']
```

```python
# New Column Names
column_names = ['CountryCode','CountryName','HultRegion','CoolName','AccessElectricity','AccessElectricityRural',
                'AccessElectricityUrban','AdolescenceFertilityRate','AgeDependency','AgeDependencyOld',
                'AgeDependencyYoung','AgricultureValueAdd','ArmyPersonnel','BirthRate','BirthsSkilled',
                'DeathRate','EducationDoctCumulative','EducationBachelorCumulative','EducationMasterCumulative',
                'EducationLowerSecondaryCumulative','EducationPostSecondCumulative','EducationPrimaryCumulative',
                'EducationTertiaryCumulative','EducationUpperSecondCumulative','EmploymentAgriculture',
                'EmploymentIndustry','EmploymentService','FertilityRate','GDP','GDPgrowth','GINI',
                'GovExpEducation','IncomeFourthQuantile','IncomeHighestQuantile','IncomeLowestQuantile',
                'IncomeSecondQuantile','IncomeThirdQuantile','IndustryValueAdd','LifeExp','LiteracyRateAdult',
                'LiteracyRateYouth','MerchandiseTrade','MilitaryExp','CellularSubscription','BelowPowertyLine',
                'PopulationYoung','PopulationMiddle','PopulationOld','PopulationDensity','PopulationGrowth',
                'PopulationLargestCity','PopulationSlums','PopulationFemale','PopulationMale',
                'PopulationTotal','PovertyHeadcount1.90','PovertyHeadcount3.20','PovertyHeadcountNational',
                'HIV','Undernourishment','Underweight','PopulationRural','ServiceValueAdd','SurfaceArea',
                'TaxRevenue','PopulationUrban','PopulationUrbangrowth']
df_aus.columns = column_names
df_aus.head()



```

```python
# Descriptive Statistics of all columns (incl. categorical values)
df_aus.describe(include = 'all')
```

```python
# Displaying information overview on dataset incl. datatypes
df_aus.info()
```

```python
# Create Heatmap of the covariance of all columns
df_aus_corr = df_aus.corr().round(2)
plt.figure(figsize=(40, 40))
sns.heatmap(data       = df_aus_corr,
            cmap       = 'Blues',
            square     = True,
            annot      = True,
            linecolor  = 'black',
            linewidths = 0.5,
            cbar       = True)
plt.show()
```

```python
# Subsetting first 4 vountries (Assigned to Max in first step)

df_set_1 = df_aus.loc[['American Samoa','Australia','Fiji','French Polynesia']]

# Excel export
#df_set_1.to_excel('Countries Max.xlsx') 

# Calculating the number of missing value per country in subset 1
df_set_1.loc[['American Samoa','Australia','Fiji','French Polynesia']].isnull().sum(axis=1)
df_aus.isnull().sum(axis=1)
```

```python
# Subsetting data on Population

df_pop = df_aus[['PopulationTotal','PopulationRural','PopulationUrban','PopulationYoung','PopulationMiddle','PopulationOld']]

# Creating Correlation Matrix of population subset
df_pop_corr = df_pop.corr().round(2)

#df_pop[['PopulationTotal']].boxplot()
#df_pop_corr.to_excel('Correlation Matrix Population.xlsx')
# Create Heatmap of the covariance of all columns


df_pop_corr = df_pop.corr().round(2)
#plt.figure(figsize=(10, 10))
sns.heatmap(data       = df_pop_corr,
            cmap       = 'coolwarm',
            square     = True,
            annot      = True,
            linecolor  = 'black',
            linewidths = 0.5,
            cbar       = True)
plt.show()
```

```python
# determining the sum of missing values per column
missing_values = df_aus.isnull().sum()

# calculating percentage of missing value per column
missing_values_perc = missing_values/len(df_aus['CountryCode'])*100
df_missing_values_perc = pd.DataFrame(missing_values_perc)
df_missing_values_perc.columns = ['Number_Missing_Values']


#plotting new df containing the number of missing values per columns
plt.figure(figsize = (8,25))
sns.set(style='whitegrid')
sns.barplot(data = df_missing_values_perc, x = 'Number_Missing_Values', y = df_missing_values_perc.index, color = 'blue')
#plt.savefig('Missing_Values.png')
```

```python
#Plotting

plt.figure(figsize = (30,20))
sns.set(style='whitegrid')

# Creating subplots for the share of population ranges(Old, Middle, Young) per country
plt.subplot(3,1,1)

sns.barplot(data = df_aus, x = 'CountryName', y = 'PopulationYoung', color = 'green')
                                                    
plt.subplot(3,1,2)
sns.barplot(data = df_aus, x = 'CountryName', y = 'PopulationMiddle', color = 'blue')

plt.subplot(3,1,3)
sns.barplot(data = df_aus, x = 'CountryName', y = 'PopulationOld', color = 'red')
```
